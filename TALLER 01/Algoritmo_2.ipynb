## ALGORITMO 2


```python
# Ejercicio 2
vect = [6,3,4,5,1,55,2,98]
print("Vector original", vect)

def metodo_burbuja(a):
    n = len(a)
    for i in range(n):
        intercambio = False
        for j in range(1, n - i):
            if a[j] < a[j - 1]:
                a[j], a[j - 1] = a[j - 1], a[j]
                intercambio = True
        if not intercambio:
            break

metodo_burbuja(vect)
print("Vector ordenado por el metodoburbuja", vect)
```

## ALGORITMO 2 PARA EL QUIZIZZ

```python
arreglo = [6,3,4,5,1,55,2,98]
print("Arreglo original:", arreglo)

def metodo_burbuja(a):
    n = len(a)
    comparaciones = 0  
    for i in range(n):
        intercambio = False
        for j in range(1, n - i):
            comparaciones += 1
            if a[j] < a[j - 1]:
                a[j], a[j - 1] = a[j - 1], a[j]
                intercambio = True
        if not intercambio:
            break
    print("Numero de comparaciones:", comparaciones)

metodo_burbuja(arreglo)
print("Arreglo ordenado por el metodo burbuja:", arreglo)
print("")
arreglo = [5,4,3,2,1]
print("Arreglo inicial del Quizizz:", arreglo)
metodo_burbuja(arreglo)
print("Arreglo ordenado del Quizizz:", arreglo)
```
