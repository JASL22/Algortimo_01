# PARTICIPACIÓN EN CLASE


```python
def iterative(n):
    if n == 0:
        return 0
    else:
        x = 0
        y = 1
        for i in range(1, n):
            z = x + y
            x = y
            y = z
        return y

n = 11
y = iterative(n)
print(f"n = {n}, y = {y}")
```

## COMPLEJIDAD PRIMER CASO


```python
import time
import math
def alg_03(n):
    if n == 0:
        return 0
    else:
        x = 0
        y = 1
        for i in range(1, n):
            z = x + y
            x = y
            y = z
        return y
    
def alg_03_02(n):
    sqrt5 = math.sqrt(5)
    phi = (1 + sqrt5) / 2
    return int(round((phi ** n) / sqrt5))
# Medicion de tiempo para alg_03
start_t = time.time()
[alg_03(x) for x in range(500)]
end_t = time.time()
print("alg_03:", end_t - start_t, "segundos, no muy optimo")
# Medicion de tiempo para alg_03_02
start_t = time.time()
[alg_03_02(x) for x in range(500)]
end_t = time.time()
print("alg_03_02:", end_t - start_t, "segundos, mas optimizado")
```

## ROUNDING


```python
import math

def FL_rounding(num: float, digits: int) -> float:
    ''' Redondea el número ``num`` con la cantidad de dígitos significativos ``digits``. '''
    num = float(f"{num:.{digits - 1}e}")
    return num
def encontrar_iteracion_con_error_minimo(tol=1e-6, digits=4):
    gr = (1 + math.sqrt(5)) / 2  
    gr = FL_rounding(gr, digits)
    x = FL_rounding(0, digits)
    y = FL_rounding(1, digits)
    i = 1
    while i < 6:
        z = FL_rounding(x + y, digits)
        if y != 0:
            ratio = FL_rounding(z / y, digits)
            error = abs((gr - ratio) / gr)
        else:
            ratio = 0
            error = float('inf')
        x = y
        y = z
        i += 1
    z = FL_rounding(x + y, digits)
    if y != 0:
        ratio = FL_rounding(z / y, digits)
        error = abs((gr - ratio) / gr)
    else:
        ratio = 0
        error = float('inf')
    print(f"La iteración tiene:")
    print(f"  y(i) = {y}")
    print(f"  y(i+1) = {z}")
    print(f"  ratio = {ratio}")
    print(f"  error = {error:.6f}")
    return 6
# Ejemplo de uso
encontrar_iteracion_con_error_minimo()
```
