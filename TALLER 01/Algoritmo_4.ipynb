```python
def serie_armonica(n):
    suma = 0.0
    for i in range(1, n + 1):
        suma += 1 / i
    return suma

# Ejemplo: calcular la suma de los primeros 100000 términos
n = 100000
resultado = serie_armonica(n)
print(f"Suma de la serie armónica hasta n={n} es: {resultado}")
```
